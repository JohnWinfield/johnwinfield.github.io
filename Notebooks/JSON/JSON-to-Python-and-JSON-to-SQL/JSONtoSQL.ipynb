# Converting JSON into SQL

```python
# Imports
import pandas as pd, json, sqlite3, requests
```

```python
# Download the JSON data from open sourced github page
url = 'https://raw.githubusercontent.com/benoitvallon/100-best-books/master/books.json'

# Checking if request went through correctly
json_data = requests.get(url)
json_data.raise_for_status
```

```python
# Load JSON data into a Python variable.
bookData = json.loads(json_data.text) 
```

```python
# print(bookData)
print(json.dumps(bookData, indent=4))
```

```python
# Create a DataFrame From the JSON Data
bookdf = pd.DataFrame(bookData)
```

```python
bookdf.head()
```

```python
#  SQL Connection
conn = sqlite3.connect("books.db")
c = conn.cursor()
```

```python
# We can now convert our JSON data from a dataframe into the sqlite format.
# View SQlite files in this webviewer: http://inloop.github.io/sqlite-viewer/
bookdf.to_sql("books_table_example",conn)
```
