# Converting JSON into Python Dictionary

```python
# Imports
import json, requests
```

```python
# User Input
zipcode = input('Zipcode: ')
countrycode = input('Country Code: (us,uk,etc.): ' )

# Download the JSON data from OpenWeatherMap.org's API based on user's zipcode and countrycode
api_address = 'https://api.openweathermap.org/data/2.5/weather?zip=%s,%s&appid=d825c979bc3bb083e75ad49b8cd584e2'
url = api_address % (zipcode, countrycode)

# Checking if request went through correctly
json_data = requests.get(url)
json_data.raise_for_status
```

```python
# Load JSON data into a Python variable.
weatherData = json.loads(json_data.text) 
```

```python
# print(weatherData)
print(json.dumps(weatherData, indent=4))
```

```python
# 'weather' contains a list, and one dictionary.
# Therefore we want to pass [0] because that is the first dicitionary inside the 'weather' list.
# Then we pass the keyname 'main', to retrieve its value pair "Clouds"

formatted_data = weatherData['weather'][0]['main']
print('Today there were be ' + formatted_data)
```

```python
# Cities in a circle API
# This API returns data from cities within a definite circle that is specified by center point (lat, lon) 
# and expected number of cities (cnt) around this point.

# User Input
lat = input('lat (i.e. 55.5): ')
lon = input('lat (i.e. 37.5): ')
cnt = input('Count of cities(i.e. 10): ')
unitsOfMeasure = input('Units of Measure (i.e. imperial, metric): ')
                       
# Download the JSON data from OpenWeatherMap.org's API based on user's input of lat, lon, and cnt of cities
api_address = 'http://api.openweathermap.org/data/2.5/find?lat=%s&lon=%s&cnt=%s&units=%s&appid=d825c979bc3bb083e75ad49b8cd584e2'
url = api_address % (lat, lon, cnt, unitsOfMeasure)

# Checking if request went through correctly
json_data = requests.get(url)
json_data.raise_for_status
```

```python
# Load JSON data into a Python variable.
weatherData2 = json.loads(json_data.text) 
```

```python
# print(weatherData)
print(json.dumps(weatherData2, indent=4))
```

```python
# Printing the weather for each City (Looping through the dictionary)
i = 0
while i < int(cnt):
    # Declaring Variables
    location = weatherData2['list'][i]['name']
    temp = weatherData2['list'][i]['main']['temp']
    feels_like = weatherData2['list'][i]['main']['feels_like']
    desc = weatherData2['list'][i]['weather'][0]['description']
    
    print('The Current weather in ' + location + ": ")
    print('The temperature is ' + str(temp) + ', but it really feels like ' + str(feels_like) +' degrees.')
    print('Today is described as '+ desc + '.')
    print()
    i += 1
```
